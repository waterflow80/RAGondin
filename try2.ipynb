```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name = "thenlper/gte-small", 
    model_kwargs = {"device": "cpu"},
    encode_kwargs = {"normalize_embeddings": True}
)
```

```python
path  ="./app/upload_dir/diversite-organisation.pdf"
pages = PyPDFLoader(path).load()
```

```python
pages
```

```python
from langchain_experimental.text_splitter import SemanticChunker

text_splitter = SemanticChunker(
    embeddings=embeddings, buffer_size=3, 
    breakpoint_threshold_type='percentile', 
    min_chunk_size=1000
)
```

```python
chunks = text_splitter.create_documents(
    [''.join(p.page_content for p in pages)]
)
```

```python
chunks[1].page_content
```

```python
for semantic_chunk in chunks:
    first_sent = semantic_chunk.page_content.split(".", maxsplit=1)[0]
    for page in pages:
        if first_sent in page.page_content:
            semantic_chunk.metadata = page.metadata
            print(page.metadata)
            break
```
